```python
className = seasonid_data['className'].values.tolist()
new_class=[]
for classes in className:
    x = classes.split('(')[0]
    x = x.replace(" ","")
    new_class.append(x)
seasonid_data['className']=new_class
seasonid_data.drop('seasonImg',axis=1,inplace=True)
```

```python
class_test = seasonid_data[seasonid_data['className']=='ICON']['seasonId'].to_string().split()[1]
class_test
```

```python
bestplayers = pd.read_csv("players_data_221212.csv",encoding='cp949')
bestplayers = bestplayers.replace({'LIVE':'22LIVE','21PLA':'21PL'})

season_list = bestplayers['클래스'].values.tolist()
new_season=[]
for season in season_list:
    classid = seasonid_data[seasonid_data['className']==season]['seasonId'].to_string().split()[1]
    new_season.append(classid)
    
bestplayers['seasonId']=new_season
bestplayers.drop('포지션',axis=1,inplace=True)
```

```python
bestplayers = bestplayers.drop_duplicates()
bestplayers
```

```python
playersList = bestplayers['이름'].values.tolist()
```

```python
spid_url = requests.get('https://static.api.nexon.co.kr/fifaonline4/latest/spid.json',headers=headers)
spid_parsed_data = spid_url.json()
spid_data = pd.DataFrame(spid_parsed_data,index=[i for i in range(len(spid_parsed_data))])
spid_data
```

```python
playersId=[]
for players in playersList:
    spid = spid_data[spid_data['name']==players]
    spid_id = spid['id'].values.tolist()
    x = str(spid_id[0])[3:]
    playersId.append(x)

id_list=[]
seasonId = bestplayers['seasonId'].values.tolist()
for players,season in zip(playersId,seasonId):
    ID = season+players
    id_list.append(ID)
    
bestplayers['ID']=id_list
bestplayers
```

```python
position=[]
index_list = bestplayers.index.tolist()
for index in index_list:
    if index>=0 and index<=15:
        position.append('ST')
    elif index>=16 and index<=31:
        position.append('CF')
    elif index>=32 and index<=47:
        position.append('LW')
    elif index>=48 and index<=63:
        position.append('RW')
    elif index>=64 and index<=79:
        position.append('CM')
    elif index>=80 and index<=95:
        position.append('CAM')
    elif index>=96 and index<=111:
        position.append('CDM')
    elif index>=112 and index<=127:
        position.append('LM')
    elif index>=128 and index<=143:
        position.append('RM')
    elif index>=144 and index<=159:
        position.append('CB')
    elif index>=160 and index<=175:
        position.append('LB')
    elif index>=176 and index<=191:
        position.append('RB')
    elif index>=192 and index<=207:
        position.append('LWB')
    elif index>=208 and index<=223:
        position.append('RWB')
    elif index>=224 and index<=239:
        position.append('GK')

bestplayers['position']=position

positionNumber=[]
for p in position:
    if p=='ST':
        positionNumber.append(25)
    elif p=='CF':
        positionNumber.append(21)
    elif p=='LW':
        positionNumber.append(27)
    elif p=='RW':
        positionNumber.append(23)
    elif p=='CM':
        positionNumber.append(14)
    elif p=='CAM':
        positionNumber.append(18)
    elif p=='CDM':
        positionNumber.append(10)
    elif p=='LM':
        positionNumber.append(16)
    elif p=='RM':
        positionNumber.append(12)
    elif p=='CB':
        positionNumber.append(5)
    elif p=='LB':
        positionNumber.append(7)
    elif p=='RB':
        positionNumber.append(3)
    elif p=='LWB':
        positionNumber.append(8)
    elif p=='RWB':
        positionNumber.append(2)
    elif p=='GK':
        positionNumber.append(0)

bestplayers['positionNumber']=positionNumber

bestplayers.to_csv('players_info.csv',encoding='cp949')
```
