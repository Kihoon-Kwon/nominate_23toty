```python
# 피온4 데일리차트 홈페이지 접속하기 
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
import chromedriver_autoinstaller
import sys
import os
import re
import requests
import pandas as pd
import numpy as np
import time

chrome_path = chromedriver_autoinstaller.install()
driver = webdriver.Chrome(chrome_path)
url = 'https://fifaonline4.nexon.com/datacenter/dailysquad'
driver.get(url)
response = requests.get(url)
html = response.text
```

```python
best_players = []

move = driver.find_element(By.XPATH,'//*[@id="positionSlider"]/div/span')
time.sleep(3)
for i in range(2,17): # 총 15개의 포지션별 인기 선수를 수집
    positions = []
    position = driver.find_element(By.CSS_SELECTOR,'#middle > div > div > div.daily > div.position_pl.daily_wrap > div.position_pl__cont.content_wrap > div.position_tab > div:nth-child('+str(i)+') > a').text
    time.sleep(4)
    select = driver.find_element(By.CSS_SELECTOR,'#middle > div > div > div.daily > div.position_pl.daily_wrap > div.position_pl__cont.content_wrap > div.position_tab > div:nth-child('+str(i)+') > a').click()
    positions.append(position)
    time.sleep(8)
    top1_SeasonAndName = driver.find_element(By.CSS_SELECTOR,'#divBestUsePositionPlayer > div.item_list.swiper-slide.swiper-slide-active > div > div > div.info > span.name_wrap')
    time.sleep(8)
    top1_divTag = top1_SeasonAndName.find_element(By.CSS_SELECTOR,'span.icon > img')
    time.sleep(8)
    top1_src = top1_divTag.get_attribute('src')
    time.sleep(8)
    top1_regex = re.compile('{}(.*){}'.format(re.escape('season/'), re.escape('.png')))
    top1_text = top1_regex.findall(str(top1_src))
    top1_name = top1_SeasonAndName.text
    top1_season_name = top1_text[0]+' '+top1_name
    positions.append(top1_season_name) # 포지션별 1위는 css 양식이 달라 따로 수집해서 append 하였음
    for j in range(2,10):
        SeasonAndName = driver.find_element(By.CSS_SELECTOR,'#divBestUsePositionPlayer > div:nth-child('+str(j)+') > div > div > div.info > span.name_wrap')
        time.sleep(8)
        divTag = SeasonAndName.find_element(By.CSS_SELECTOR,'span.icon > img')
        time.sleep(8)
        src = divTag.get_attribute('src')
        time.sleep(8)
        regex = re.compile('{}(.*){}'.format(re.escape('season/'), re.escape('.png')))
        text = regex.findall(str(src)) # 선수 시즌 추출(리스트 형식이므로 원소 추출하는 방식으로 데이터 가져와야 됨)
        name = SeasonAndName.text # 선수 이름 추출
        season_name = text[0]+' '+name 
        positions.append(season_name)
        time.sleep(8)
    
    ActionChains(driver).drag_and_drop_by_offset(move,549,0).perform() # 스크롤 오른쪽으로 이동

    for k in range(10,17):
        SeasonAndName = driver.find_element(By.CSS_SELECTOR,'#divBestUsePositionPlayer > div:nth-child('+str(k)+') > div > div > div.info > span.name_wrap')
        time.sleep(5)
        divTag = SeasonAndName.find_element(By.CSS_SELECTOR,'span.icon > img')
        src = divTag.get_attribute('src')
        regex = re.compile('{}(.*){}'.format(re.escape('season/'), re.escape('.png')))
        text = regex.findall(str(src))
        name = SeasonAndName.text
        season_name = text[0]+' '+name
        positions.append(season_name)
        
    best_players.append(positions)
    print('process {0} complete'.format(i-1))
    print(positions)
    print('------------------------------------------------------------------------------------')
```

# 데이터 시각화

- 선수의 포지션,이름,클래스를 데이터프레임으로 시각화한다. 포지션은 선수 카드에 주어진 포지션이 아니라 유저들이 사용한 포지션을 따른다.
   
   - ex) TKL 차두리의 원래 포지션은 RB이지만, 유저들은 CB으로 주로 사용하므로 TKL 차두리는 CB로 분류
    
- 이후 csv 파일로 데이터프레임을 저장 후, 자동 스쿼드 제작에 활용한다.   

```python
players_dataset = []
columns = ['포지션','이름','클래스']
best_player = best_players[0] # 굳이 0 아니어도 됨, 포지션별 16명이므로 하나 정해두고
for i in range(len(best_players)):
    for k in range(len(best_player)-1):
        players_dataset.append([best_players[i][0],best_players[i][k+1].split(' ',1)[1],best_players[i][k+1].split(' ',1)[0]])
        
players_df = pd.DataFrame(players_dataset,columns=columns)
players_df # 시각화 완료
```
