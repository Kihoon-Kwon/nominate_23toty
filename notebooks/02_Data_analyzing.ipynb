```python
# 1-1. 공격력 요인의 가중치

GoalWeight = 0.625
AssistWeight = 0.375

# 1-2. 공격력에 대한 포지션별 가중치
FW_GoalWeight = 0.161
MF_GoalWeight = 0.285
DF_GoalWeight = 0.554
FW_AssistWeight = 0.208
MF_AssistWeight = 0.284
DF_AssistWeight = 0.508
```

```python
# 공격력 평점 구하는 법 → (골 기록*골 가중치*포지션별 골 가중치)+(도움 기록*도움 가중치*포지션별 도움 가중치)

FW_goal_per = df_forward['경기 당 골 횟수'].tolist()
FW_assist_per = df_forward['경기 당 도움 횟수'].tolist()

FW_attack = []
for goal,assist in zip(FW_goal_per,FW_assist_per):
    score = (goal*GoalWeight*FW_GoalWeight)+(assist*AssistWeight*FW_AssistWeight)
    FW_attack.append(score*100)
    
MF_goal_per = df_midfielder['경기 당 골 횟수'].tolist()
MF_assist_per = df_midfielder['경기 당 도움 횟수'].tolist()

MF_attack = []
for goal,assist in zip(MF_goal_per,MF_assist_per):
    score = (goal*GoalWeight*MF_GoalWeight)+(assist*AssistWeight*MF_AssistWeight)
    MF_attack.append(score*100)
    
DF_goal_per = df_defender['경기 당 골 횟수'].tolist()
DF_assist_per = df_defender['경기 당 도움 횟수'].tolist()

DF_attack = []
for goal,assist in zip(DF_goal_per,DF_assist_per):
    score = (goal*GoalWeight*DF_GoalWeight)+(assist*AssistWeight*DF_AssistWeight)
    DF_attack.append(score*100)
    
df_forward['공격력']=FW_attack
df_midfielder['공격력']=MF_attack
df_defender['공격력']=DF_attack
```

```python
# 2-1 기본능력 요인의 가중치

ShootWeight = 0.348
PassWeight = 0.421
DribbleWeight = 0.231

# 2-2 기본능력에 대한 포지션별 가중치

FW_ShootWeight = 0.129
FW_PassWeight = 0.495
FW_DribbleWeight = 0.542
MF_ShootWeight = 0.277
MF_PassWeight = 0.307
MF_DribbleWeight = 0.296
DF_ShootWeight = 0.594
DF_PassWeight = 0.208
DF_DribbleWeight = 0.162
```

```python
# 기본능력 평점 구하는 법 → (슈팅 기록*슈팅 가중치*포지션별 슈팅 가중치)+
#                            (패스 기록*패스 가중치*포지션별 패스 가중치)+
#                            (드리블 기록*드리블 가중치*포지션별 드리블 가중치)

FW_shoot_per = df_forward['경기 당 슈팅 횟수'].tolist()
FW_pass_per = df_forward['경기 당 패스 횟수'].tolist()
FW_dribble_per = df_forward['경기 당 드리블 횟수'].tolist()

FW_ability = []
for shoot,passes,dribble in zip(FW_shoot_per,FW_pass_per,FW_dribble_per):
    score = (shoot*ShootWeight*FW_ShootWeight)+(passes*PassWeight*FW_PassWeight)+(dribble*DribbleWeight*FW_DribbleWeight)
    FW_ability.append(score*100)
    
MF_shoot_per = df_midfielder['경기 당 슈팅 횟수'].tolist()
MF_pass_per = df_midfielder['경기 당 패스 횟수'].tolist()
MF_dribble_per = df_midfielder['경기 당 드리블 횟수'].tolist()

MF_ability = []
for shoot,passes,dribble in zip(MF_shoot_per,MF_pass_per,MF_dribble_per):
    score = (shoot*ShootWeight*MF_ShootWeight)+(passes*PassWeight*MF_PassWeight)+(dribble*DribbleWeight*MF_DribbleWeight)
    MF_ability.append(score*100)
    
DF_shoot_per = df_defender['경기 당 슈팅 횟수'].tolist()
DF_pass_per = df_defender['경기 당 패스 횟수'].tolist()
DF_dribble_per = df_defender['경기 당 드리블 횟수'].tolist()

DF_ability = []
for shoot,passes,dribble in zip(DF_shoot_per,DF_pass_per,DF_dribble_per):
    score = (shoot*ShootWeight*DF_ShootWeight)+(passes*PassWeight*DF_PassWeight)+(dribble*DribbleWeight*DF_DribbleWeight)
    DF_ability.append(score*100)
    
df_forward['기본 능력']=FW_ability
df_midfielder['기본 능력']=MF_ability
df_defender['기본 능력']=DF_ability
```

```python
# 3-1 수비력 요인의 가중치

TackleWeight = 0.308
InterceptWeight = 0.311
ClearingWeight = 0.205
BlockWeight = 0.176

# 3-2 수비력에 대한 포지션별 가중치

FW_TackleWeight = 0.459
FW_InterceptWeight = 0.473
FW_ClearingWeight = 0.442
FW_BlockWeight = 0.446
MF_TackleWeight = 0.273
MF_InterceptWeight = 0.328
MF_ClearingWeight = 0.304
MF_BlockWeight = 0.296
DF_TackleWeight = 0.268
DF_InterceptWeight = 0.199
DF_ClearingWeight = 0.254
DF_BlockWeight = 0.258
```

```python
# 수비력 평점 구하는 법 → (태클 기록*태클 가중치*포지션별 태클 가중치)+
#                          (인터셉트 기록*인터셉트 가중치*포지션별 인터셉트 가중치)+
#                          (클리어링 기록*클리어링 가중치*포지션별 클리어링 가중치)+
#                          (블락 기록*블락 가중치*포지션별 블락 가중치)

FW_tackle_per = df_forward['경기 당 태클 횟수'].tolist()
FW_intercept_per = df_forward['경기 당 가로채기 횟수'].tolist()
FW_clearing_per = df_forward['경기 당 클리어링 횟수'].tolist()
FW_block_per = df_forward['경기 당 차단 횟수'].tolist()

FW_defense = []
for tackle,intercept,clearing,block in zip(FW_tackle_per,FW_intercept_per,FW_clearing_per,FW_block_per):
    score = (tackle*TackleWeight*FW_ShootWeight)+(intercept*InterceptWeight*FW_InterceptWeight)+(clearing*ClearingWeight*FW_ClearingWeight)+(block*BlockWeight*FW_BlockWeight)
    FW_defense.append(score*100)
    
MF_tackle_per = df_midfielder['경기 당 태클 횟수'].tolist()
MF_intercept_per = df_midfielder['경기 당 가로채기 횟수'].tolist()
MF_clearing_per = df_midfielder['경기 당 클리어링 횟수'].tolist()
MF_block_per = df_midfielder['경기 당 차단 횟수'].tolist()

MF_defense = []
for tackle,intercept,clearing,block in zip(MF_tackle_per,MF_intercept_per,MF_clearing_per,MF_block_per):
    score = (tackle*TackleWeight*MF_ShootWeight)+(intercept*InterceptWeight*MF_InterceptWeight)+(clearing*ClearingWeight*MF_ClearingWeight)+(block*BlockWeight*MF_BlockWeight)
    MF_defense.append(score*100)
    
DF_tackle_per = df_defender['경기 당 태클 횟수'].tolist()
DF_intercept_per = df_defender['경기 당 가로채기 횟수'].tolist()
DF_clearing_per = df_defender['경기 당 클리어링 횟수'].tolist()
DF_block_per = df_defender['경기 당 차단 횟수'].tolist()

DF_defense = []
for tackle,intercept,clearing,block in zip(DF_tackle_per,DF_intercept_per,DF_clearing_per,DF_block_per):
    score = (tackle*TackleWeight*DF_ShootWeight)+(intercept*InterceptWeight*DF_InterceptWeight)+(clearing*ClearingWeight*DF_ClearingWeight)+(block*BlockWeight*DF_BlockWeight)
    DF_defense.append(score*100)
    
df_forward['수비력']=FW_defense
df_midfielder['수비력']=MF_defense
df_defender['수비력']=DF_defense
```

```python
# 4-1 기여도 요인의 가중치(공중볼 경합만 반영, 세컨볼 획득에 대한 데이터는 없음)

FW_headerWeight = 0.392
MF_headerWeight = 0.258
DF_headerWeight = 0.351
```

```python
# 기여도 요인 평점 구하는 법 → (공중볼 경합 기록*포지션별 공중볼 경합 가중치)

FW_header_per = df_forward['경기 당 공중볼 경합 횟수']
FW_contribution = []
for header in FW_header_per:
    score = header*FW_headerWeight
    FW_contribution.append(score*100)

MF_header_per = df_midfielder['경기 당 공중볼 경합 횟수']
MF_contribution = []
for header in MF_header_per:
    score = header*MF_headerWeight
    MF_contribution.append(score*100)
    
DF_header_per = df_defender['경기 당 공중볼 경합 횟수']
DF_contribution = []
for header in DF_header_per:
    score = header*DF_headerWeight
    DF_contribution.append(score*100)
    
df_forward['기여도']=FW_contribution
df_midfielder['기여도']=MF_contribution
df_defender['기여도']=DF_contribution
```

```python
df_forward['총점'] = df_forward['공격력']+df_forward['기본 능력']+df_forward['수비력']+df_forward['기여도']
df_midfielder['총점'] = df_midfielder['공격력']+df_midfielder['기본 능력']+df_midfielder['수비력']+df_midfielder['기여도']
df_defender['총점'] = df_defender['공격력']+df_defender['기본 능력']+df_defender['수비력']+df_defender['기여도']
```
